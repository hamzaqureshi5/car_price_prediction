```python
data = pd.read_csv("CarPrice_Assignment.csv")
data.head()
```

```python
# data.isnull().sum()
data.drop(columns = ['car_ID'], inplace= True)
```

```python
len(data['CarName'].unique())
```

```python
final_car = pd.get_dummies(data, columns=['symboling','CarName','fueltype',	'aspiration',	'doornumber',	'carbody',	'drivewheel',	'enginelocation',	'enginetype',	'cylindernumber',	'fuelsystem'])
final_car
```

```python
final_car.describe()
```

```python
import numpy as np
#for seperating 70 percent of data
np.random.seed(11111)
msk = np.random.rand(len(final_car)) < 0.70
train_total = final_car[msk]
test_total = final_car[~msk]

```

```python
train_total["price"]
```

```python
train_label = train_total.loc[:,'price']
test_label = test_total.loc[:,'price']
```

```python
test_data  = test_total.drop(columns = ['price'])
train_data = train_total.drop(columns = ['price'])
```

```python
test_data
```

```python
#Normalization
mean = train_data.iloc[:,0:13].mean(axis=0)
train_data.iloc[:,0:13] -= mean
std = train_data.iloc[:,0:13].std(axis=0)
train_data.iloc[:,0:13] /= std
test_data.iloc[:,0:13] -= mean
test_data.iloc[:,0:13] /= std

```

```python
mean_label = train_label.mean()
train_label -= mean_label
std_label = train_label.std()
train_label /= std_label
test_label -= mean_label
test_label /= std_label
```

```python
#data converted to array
test = np.array(test_data.iloc[:])
train = np.array(train_data.iloc[:])
test_l= np.array(test_label.astype('float32'))
train_l= np.array(train_label.astype('float32'))
```

```python
! pip install scikit-learn
import datetime

import numpy as np
import pandas as pd

# import matplotlib.pyplot as plt
# import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
linearRegression = LinearRegression()
linearRegression.fit(train, train_l)
```

```python
y_pred = linearRegression.predict(test)
#y_pred
r2_score(test_l, y_pred)
```

```python
rf = RandomForestRegressor(n_estimators = 100)
rf.fit(train, train_l)
y_pred = rf.predict(test)
r2_score(test_l, y_pred)
```
